## Citation Class Overlap Analysis

This notebook evaluates how often patent topic–candidate pairs share at least one IPC class.


### Conclusion
Citation/retrieval pairs are mostly class-consistent (high IPC overlap), but a non-trivial minority lacks class overlap, indicating potentially broader or cross-class citation behavior. Therefore, filtering against IPC prior to the retrieval might lead to dropping some of the relevant citations

```python
import pandas as pd
import plotly.express as px
from pathlib import Path
import os
from patent_retrieval import utils as utils
```

```python
# paths to the analysis file output by `patent_class.py` and the test topics file
analysis_path = "/home/alm3rng/patent-retrieval/src/patent_retrieval/dataset/patent_classification_analysis_B00X 000.csv"
test_topics_path=Path(os.environ["CLEF_IP_LOCATION"])/ "02_topics"/ "test-pac"/ "relass_clef-ip-2011-PAC_abs.txt"
```

```python
classes_df = pd.read_csv(analysis_path)
classes_df
```

```python
classes_df["topic_ipcrs"] = classes_df["topic_ipcrs"].map(lambda x: {c.split()[0] for c in eval(x)})
classes_df["doc_ipcrs"] = classes_df["doc_ipcrs"].map(lambda x: {c.split()[0] for c in eval(x)})
```

```python
classes_df['common_ipcrs'] = classes_df.apply(
    lambda r: (r['topic_ipcrs'] or set()).intersection(r['doc_ipcrs'] or set()),
    axis=1
)

classes_df['n_common_ipcrs'] = classes_df['common_ipcrs'].apply(len)
```

```python
classes_df
```

```python
classes_df_top200 = classes_df.groupby('topic').head(100).reset_index(drop=True)
classes_df_top200
```

```python
share_counts = (
    (classes_df_top200["n_common_ipcrs"] > 0)
    .map({True: "share >=1 class", False: "share 0 classes"})
    .value_counts()
    .rename_axis("pair_type")
    .reset_index(name="count")
)

share_counts["pct"] = 100 * share_counts["count"] / share_counts["count"].sum()
share_counts["label"] = share_counts["pct"].map(lambda x: f"{x:.1f}%")

fig = px.bar(
    share_counts,
    x="pair_type",
    y="count",
    text="label",
    title="Topic–candidate pairs: at least one shared class vs none"
)
fig.update_traces(textposition="outside")
fig.update_layout(xaxis_title="", yaxis_title="Number of pairs")
fig.show()
```

```python
topics_df = utils.load_topics_df(test_topics_path)
topics_df
```

```python
topic_to_ipcrs = classes_df.groupby("topic")["topic_ipcrs"].first()
topics_df["topic_ipcrs"] = topics_df["topic"].map(topic_to_ipcrs)
candidates_to_ipcrs = classes_df.groupby("number")["doc_ipcrs"].first()
topics_df["candidate_ipcrs"] = topics_df["candidate"].map(candidates_to_ipcrs)

topics_df
```

```python
topics_df['common_ipcrs'] = topics_df.apply(
    lambda r: (
        r['topic_ipcrs'] if isinstance(r['topic_ipcrs'], set) else set()
    ).intersection(
        r['candidate_ipcrs'] if isinstance(r['candidate_ipcrs'], set) else set()
    ),
    axis=1
)
topics_df['n_common_ipcrs'] = topics_df['common_ipcrs'].apply(len)
```

```python
share_counts = (
    (topics_df["n_common_ipcrs"] > 0)
    .map({True: "share >=1 class", False: "share 0 classes"})
    .value_counts()
    .rename_axis("pair_type")
    .reset_index(name="count")
)

share_counts["pct"] = 100 * share_counts["count"] / share_counts["count"].sum()
share_counts["label"] = share_counts["pct"].map(lambda x: f"{x:.1f}%")

fig = px.bar(
    share_counts,
    x="pair_type",
    y="count",
    text="label",
    title="Topic–candidate pairs: at least one shared class vs none"
)
fig.update_traces(textposition="outside")
fig.update_layout(xaxis_title="", yaxis_title="Number of pairs")
fig.show()
```

## Summary of Insights

**Analysis 1 – Citation pairs (top-100 candidates per topic):**
- **79.8%** of topic–candidate pairs share at least one IPC class.
- **20.2%** share zero IPC classes.

**Analysis 2 – Full test-set topic–candidate pairs:**
- **85.0%** share at least one IPC class.
- **15.0%** share zero IPC classes.

**Key takeaways:**
1. The majority of citation/retrieval pairs are **class-consistent**, confirming that IPC classification is a strong signal for patent relevance.
2. However, a **non-trivial minority (15–20%)** of relevant citations fall outside the topic's IPC classes, indicating cross-class citation behavior.
3. Using IPC as a hard pre-filter for retrieval would **discard 15–20% of true relevant documents**, making it a risky strategy unless recall loss is acceptable.
4. IPC could still be useful as a **soft signal** (e.g., boosting scores for same-class candidates) rather than a strict filter.
